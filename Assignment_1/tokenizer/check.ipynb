```python
import re # python module for regular expressions
# now defining the common regular expression (regex) patterns for these 3 languages

import re

TOKEN_PATTERN = re.compile(rf"""
    https?://[^\s]+[\.\w{{2,}}] |                       # URLs
    [\w\.-]+@[\w\.-]+\.\w+ |                # Emails
    \d{{1,2}}[/-]\d{{1,2}}[/-]\d{{2,4}} |   # Dates (dd-mm-yyyy or dd/mm/yyyy)
    \d{{1,2}}:\d{{2}}(?::\d{{2}})? |        # Time (hh:mm[:ss])
    [₹$€¥]\s?\d+(?:\.\d+)? |                # Currency
    \d+(?:\.\d+)? |                         # Numbers (int, float)
    [\w\u0900-\u097F\u0A80-\u0AFF]+ |       # Words (English, Hindi, Gujarati)
    [^\w\s]                                 # Punctuation
""", re.VERBOSE | re.UNICODE)
```

```python
def sentenceTokenizer(text):
    sentence_endings = re.compile(r'(?<=[.?!।])\s+(?=[A-Z\u0900-\u097F\u0A80-\u0AFF])')
    return [s.strip() for s in sentence_endings.split(text.strip()) if s.strip()]
```

```python
def wordTokenizer(sentence):
    return TOKEN_PATTERN.findall(sentence)
```

```python
def preprocessing(example):
    text = example['text']  
    sentences = sentenceTokenizer(text)
    tokens_per_sentence = [wordTokenizer(s) for s in sentences]
    return {
        "sentences": sentences,
        "tokens": tokens_per_sentence
    }
```

```python
from datasets import load_dataset

# english dataset
en_dataset = load_dataset(
    "ai4bharat/IndicCorpV2",
    data_files="data/en.txt",
    split="train",
    streaming=True
)

en_processed = (preprocessing(example) for example in en_dataset)

for i, sample in enumerate(en_processed):
    print("English:", sample)
    if i == 2:
        break
```

```python
import json

def save_tokenized_dataset(tokenized_data, lang_code):
    output_file = f"tokenized_{lang_code}.jsonl"
    with open(output_file, "w", encoding="utf-8") as f:
        for item in tokenized_data:
            json.dump(item, f, ensure_ascii=False)
            f.write("\n")
    print(f"Tokenized data saved to {output_file}")
```

```python
def tokens_only_stream(dataset):
    for example in dataset:
        processed = preprocessing(example)   # gives {"sentences": ..., "tokens": ...}
        yield {"tokens": processed["tokens"]}
```

```python
import json
from collections import Counter

def calculate_corpus_stats(jsonl_path):
    num_documents = 0
    num_sentences = 0
    num_tokens = 0
    num_characters = 0
    sentence_lengths = []
    document_lengths = []
    vocab_counter = Counter()
    skipped_lines = 0

    with open(jsonl_path, 'r', encoding='utf-8') as f:
        for line_num, line in enumerate(f, start=1):
            try:
                data = json.loads(line)
            except json.JSONDecodeError as e:
                print(f"Skipping line {line_num} due to JSON error: {e}")
                skipped_lines += 1
                continue

            num_documents += 1
            sentences = data.get("sentences", [])
            tokens_per_sentence = data.get("tokens", [])

            num_sentences += len(sentences)

            flat_tokens = [token for sent in tokens_per_sentence for token in sent]
            num_tokens += len(flat_tokens)
            num_characters += sum(len(token) for token in flat_tokens)

            sentence_lengths.extend([len(sent) for sent in tokens_per_sentence])
            document_lengths.append(len(flat_tokens))

            vocab_counter.update(flat_tokens)

    avg_sentence_length = sum(sentence_lengths) / len(sentence_lengths) if sentence_lengths else 0
    avg_document_length = sum(document_lengths) / len(document_lengths) if document_lengths else 0
    ttl_ratio = num_tokens / num_documents if num_documents else 0
    avg_word_length = num_characters / num_tokens if num_tokens else 0
    type_token_ratio = len(vocab_counter) / num_tokens if num_tokens else 0

    return {
        "Total Documents": num_documents,
        "Total Sentences": num_sentences,
        "Total Words (Tokens)": num_tokens,
        "Total Characters": num_characters,
        "Average Sentence Length (in words)": round(avg_sentence_length, 2),
        "Average Word Length (in characters)": round(avg_word_length, 2),
        "Average Document Length (in tokens)": round(avg_document_length, 2),
        "TTL Ratio (Tokens-to-Lines)": round(ttl_ratio, 2),
        "Vocabulary Size": len(vocab_counter),
        "Type/Token Ratio (TTR)": round(type_token_ratio, 4),
        "Skipped Lines": skipped_lines
    }
```

```python
# GUJRATI
stats = calculate_corpus_stats("tokenized_gu.jsonl")
for k, v in stats.items():
    print(f"{k}: {v}")
```
