#### Creating 3 sets containing:  all symbols in python, lowercase & uppercase letters respectively

```python
from automathon import DFA
import string

sigma=set(string.printable)
lowercase=set(string.ascii_lowercase)
uppercase=set(string.ascii_uppercase)
```

#### Defining the states, start state, and accepting state

```python
q={'q0','q1','q2'}
delta={s: {} for s in q}
initial_state='q0'
f={'q1'}
```

#### Defining transitions: A dictionary that maps (current_state, input_symbol) to  next_state

```python
# for q0 

for char in sigma:
    if char in lowercase or char in uppercase:
        delta['q0'][char]='q1'
    else:
        delta['q0'][char]='q2'


# for q1, the accepting state

for char in sigma:
    if char in lowercase:
        delta['q1'][char]='q1'
    else:
        delta['q1'][char]='q2'


# for q2, the trap

for char in sigma:
    delta['q2'][char]='q2'
```

#### Building the DFA

```python
automata=DFA(q,sigma,delta,initial_state,f)
```

#### Checking if the automata is valid

```python
automata.is_valid()
```

#### Checking validity of the sample testcases

```python
print(automata.accept("Cat")) # valid from here 
print(automata.accept("cat"))
print(automata.accept("dog"))
print(automata.accept("a"))
print(automata.accept("dog1")) # not valid from here
print(automata.accept("1dog"))
print(automata.accept("DogHouse"))
print(automata.accept("Dog_house"))
print(automata.accept(" cats"))
```

```python
from automata.fa.dfa import DFA

from visual_automata.fa.dfa import VisualDFA
```

```python
dfa = VisualDFA(automata)
invalid_chars = sigma - lowercase  # Anything not a-z
```

```python
dfa = VisualDFA(
    states={"q0", "q1", "q2"},
    input_symbols=sigma,
    transitions={
        "q0": {c: "q1" for c in lowercase} | {c: "q2" for c in invalid_chars},
        "q1": {c: "q1" for c in lowercase} | {c: "q2" for c in invalid_chars},
        "q2": {c: "q2" for c in lowercase.union(invalid_chars)}
    },

    initial_state="q0",
    final_states={"q1"},
)
```

```python
dfa = VisualDFA(dfa)
```
